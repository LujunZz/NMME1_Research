```python
def NSE_Func(measurement,simulation):
    numerator = np.square(measurement - simulation)
    numerator = numerator.sum()
    denominator = np.power(measurement - measurement.mean(),2)
    denominator = denominator.sum()
    NSE = 1 - numerator/denominator
    return[NSE]
```

```python
# Get NMME members Variable List along with their dimensions
def Get_NMME_Var_Dict(ncInput):
    Var_name = list(ncInput.variables.keys())
    if np.size(Var_name) == 6:
        Var_name.remove('prec')
    else:
        Var_name.remove('prate')
    Var_Dict = {}
    for i in Var_name:
        Var_Dict[i]= np.size(ncInput[i][:])
    return Var_Dict
```

```python
def Get_NMME_index_order(ncInput):
    Var_name = str(ncInput)
    start_index = Var_name.find('prec')
    if Var_name.find('prec')!=-1:
        Var_name = Var_name[start_index+5:start_index+14]
        Order_list = ['a','a','a','a','a']
        for i in range(len(Var_name)):
            if i%2 == 0:
                Order_list[int(i/2)] = Var_name[i]               
    elif Var_name.find('prec')==-1:
        start_index = Var_name.find('prate')
        Var_name = Var_name[start_index+6:start_index+11]
        Order_list = ['a','a','a']
        for i in range(len(Var_name)):
            if i%2 == 0:
                Order_list[int(i/2)] = Var_name[i]        
    else:
        print ('Error there have no dim')
        os.system("pause")
        exit(0)
    return Order_list
```

```python
def Get_NMME_Precip(ncInput,Leadtime_idx):
    Order_list = Get_NMME_index_order(ncInput)
    if np.size(Order_list)==5 and Order_list[1] == 'L':
        Var_name = 'prec'
        varData = ncInput.variables[Var_name][:,Leadtime_idx,:,:,:]
        return varData
    if np.size(Order_list)==5 and Order_list[2] == 'L':
        Var_name = 'prec'
        varData = ncInput.variables[Var_name][:,:,Leadtime_idx,:,:]
        return varData 
    if np.size(Order_list)==3:
        Var_name = 'prate'
        varData = ncInput.variables[Var_name][:,:,:]   
        return varData
```

```python
# Clipping NMME Members Using shapefile
def NMME_Precip_Clip_Via_Shp(ncInput,ShpFileDirec,Out_Direc,Leadtime_idx):
    '''
    NcInput -- a NetCDF4.Dataset object
    ShpFileDirec -- Shape file's directory in str format
    Out_Direct -- Saving directory of clipped nc Dataset
    '''
    #read shapefile
    shpDS = ogr.Open(ShpFileDirec)
    shpLyr = shpDS.GetLayer()
    Envelop = shpLyr.GetExtent() 
    xmin,xmax,ymin,ymax = [Envelop[0],Envelop[1],Envelop[2],Envelop[3]]    #Your extents as given above
    mask_RES = []
    print('Completed Reading Data!')
    ######################################################
    #                      Process                       #
    ######################################################
    Order_List = Get_NMME_index_order(ncInput)
    Var_Dict = Get_NMME_Var_Dict(ncInput)
    var_name = list(ncInput.variables.keys())
    # To judge whether prec or prate
    judge = (np.size(var_name) == 6)
    if judge:
        ref_Ori = ncInput['S'][:]
        member_Ori = ncInput['M'][:]
    else:
        ref_Ori = ncInput['T'][:]
    lon_Ori = ncInput['X'][:]
    lat_Ori = ncInput['Y'][:]
    varData_Ori = Get_NMME_Precip(ncInput,Leadtime_idx)
    print('Completed Extracting Data!')
    ######################################################
    #                    Create mask                     #
    ######################################################
    if len(mask_RES) == 0 :
        #get boundary and xs ys
        lat_bnds, lon_bnds = [ymin, ymax], [xmin+180, xmax+180]
        lat_inds = np.where((lat_Ori >= (lat_bnds[0])-1) & (lat_Ori <= lat_bnds[1]))
        lon_inds = np.where((lon_Ori >= (lon_bnds[0])-1) & (lon_Ori <= lon_bnds[1]))
        ncols = len(lon_inds[0])
        nrows = len(lat_inds[0])
        nreftime = len(ref_Ori)
        if judge:
            nmember = len(member_Ori)
        #create geotransform
        xres = (xmax - xmin) / float(ncols)
        yres = (ymax - ymin) / float(nrows)
        geotransform = (xmin+0.5,xres,0,ymax+0.5,0,-yres)
        #create mask
        mask_DS = gdal.GetDriverByName('MEM').Create('', ncols, nrows, 1 ,gdal.GDT_Int32)
        mask_RB = mask_DS.GetRasterBand(1)
        mask_RB.Fill(0) #initialise raster with zeros
        mask_RB.SetNoDataValue(-32767)
        mask_DS.SetGeoTransform(geotransform)
        maskvalue = 1
        err = gdal.RasterizeLayer(mask_DS, [maskvalue], shpLyr)
        mask_DS.FlushCache()
        mask_array = mask_DS.GetRasterBand(1).ReadAsArray()    
        mask_RES = ma.masked_equal(mask_array, 255)          
        ma.set_fill_value(mask_RES, -32767) 
        print('Completed Creating Mask!')
    ######################################################
    #                      Subset                        #
    ######################################################
    if judge:
        var_subset = varData_Ori[:,:,min(lat_inds[0])-1:max(lat_inds[0]), min(lon_inds[0])-1:max(lon_inds[0])]
    else:
        var_subset = varData_Ori[:,min(lat_inds[0])-1:max(lat_inds[0]), min(lon_inds[0])-1:max(lon_inds[0])]
    #if np.any(mask_RES.mask):
    #    var_subset.__setmask__(np.logical_not(np.flipud(mask_RES.mask))) # update mask (flipud is reverse 180)
    lon_subset = lon_Ori[lon_inds]-180
    lat_subset = lat_Ori[lat_inds]
    print('Completed Subseting Data!')
    ######################################################
    # Open a new NetCDF file to for data svaing. You can #
    # choose one of the formats from 'NETCDF3_CLASSIC',  #
    # 'NETCDF3_64BIT', 'NETCDF4_CLASSIC', and 'NETCDF4'  #
    ######################################################
    # Using our previous dimension info, we can create the new time dimension
    # Even though we know the size, we are going to set the size to unknown
    ncOutput = Dataset(Out_Direc, 'w', format='NETCDF4')

    #ncOutput.createDimension('time', None)
    ncOutput.createDimension('X', ncols)
    ncOutput.createDimension('Y', nrows)
    if judge:
        ncOutput.createDimension('M', nmember)
    ncOutput.createDimension('S', nreftime)

    # Add lat Variable
    var_out_Y = ncOutput.createVariable('Y','f',("Y"))
    ncOutput.variables['Y'][:] = lat_subset[:]

    # Add lon Variable
    var_out_X = ncOutput.createVariable('X','f',("X"))
    ncOutput.variables['X'][:] = lon_subset[:]
    
    # Add Member Variable
    if judge:
        var_out_M = ncOutput.createVariable('M','f',("M"))
        ncOutput.variables['M'][:] = member_Ori[:]

    # Add reftime Variable
    var_out_S = ncOutput.createVariable('S','f',("S"))
    ncOutput.variables['S'][:] = ref_Ori[:]

    # Add data Variable
    if judge:
        var_out_data = ncOutput.createVariable('prec', 'f',("S",'M',"Y","X"))
        for i in range(np.size(ref_Ori)):
            ncOutput.variables['prec'][i,:,:,:] = var_subset[i,:,:,:]
    else:
        var_out_data = ncOutput.createVariable('prate', 'f',("S","Y","X"))
        for i in range(np.size(ref_Ori)):
            ncOutput.variables['prate'][i,:,:] = var_subset[i,:,:]
    print(np.shape(var_subset))
    # attr
    ncOutput.history = "CLIP Created datatime" + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") + " by LujunZ at OU"
    ncOutput.source  = "netCDF4 under python 3.6.5"
    ######################################################
    #                   Write close                      #
    ######################################################
    ncOutput.close()  # close the new file
    print('Completed Saving NcFile!')
    
    return
```

```python
# Averaging for certain member of NMME 
def Averaging_NMME_Member(ncInput,leadtime_index):
    Var_List = Get_NMME_Var_List(ncInput)
    size_Member = Var_List['M']
    for i in range(size_Member):
        if Var_List[1] == 'M':
            if i == 0:
                nc_arr = np.array(ncInput['prec'][:,i,leadtime_index,:,:],dtype='float16')
            else:
                nc_arr = nc_arr + np.array(ncInput['prec'][:,i,leadtime_index,:,:],dtype='float16')
        elif Var_List[2] =='M':
            if i == 0:
                nc_arr = np.array(ncInput['prec'][:,i,leadtime_index,:,:],dtype='float16')
            else:
                nc_arr = nc_arr + np.array(ncInput['prec'][:,i,leadtime_index,:,:],dtype='float16')
    nc_arr = nc_arr/len(ncInput['M'][:])
    return nc_arr
```

```python
# Bayesian Moedl Averagin weights computation
def Bayesian_weights(simulations, observations):
    """
    m ensemble members
    n number of observations
    simulations - m by n matrix of simulations from all members
    observations - n by 1 matrix of the observation data
    BIC - Bayesian Information Critirion
    """
    [n,m] = np.shape(simulations)
    BIC = np.zeros(m)
    wgt = np.zeros(m)
    for i in range(m):
        error_variance = np.sum(np.power(simulations[:,i]-observations,2))/n
        BIC[i] = 2*n*np.log(error_variance)
    for i in range(m):
        BIC_sub = BIC - BIC[i]
        wgt[i] = 1/np.sum(np.exp(-0.5*BIC_sub))
    return wgt
```

```python
# Computing the Bayesian weights for different members at each locations
# Calculating the Bayesian Model Averaging at each loceations
# Saving the corresponding results
for i in range(np.size(City_Name)):
    Out_Direc = 'BMA/'+City_Name[i]+'_BMA.xlsx'
    exec('data = data_'+City_Name[i]+'.drop(["RefData"],axis=1)')
    exec('data_'+City_Name[i]+'["SMA"]=data.mean(1)')
    exec('ref = data_'+City_Name[i]+'["RefData"]')
    ref_inp = np.array(ref["19820101":"19991231"])
    data_inp = np.array(data["19820101":"19991231"])
    exec('weights_'+City_Name[i]+'=Bayesian_weights(data_inp,ref_inp)')
    exec('data_'+City_Name[i]+'["BMA"] = data_'+City_Name[i]+'\
         .drop(["RefData","SMA"],axis=1).values.dot(np.reshape(weights_'+City_Name[i]+',-1,1))')
    exec('data_'+City_Name[i]+'.to_excel(Out_Direc, sheet_name="Sheet_name_1")') 
```
