```python
import os
import netCDF4
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 
%matplotlib inline
```

```python
def NC_PLOT(NC,Longname, Latname, Varname):
    Long = Longname
    Long = nc.variables[Long]
    Lat = Latname
    Lat = nc.variables[Lat][:]
    Var = Varname
    Var = nc.variables[Var][:]
    
    lon_min=np.min(Long)-2; lon_max=np.max(Long)+2
    lat_min=np.min(Lat)-2; lat_max=np.max(Lat)+2
    plt.figure(figsize=(16,9))
    m=Basemap(projection='cyl',llcrnrlon=lon_min,urcrnrlon=lon_max,llcrnrlat=lat_min,urcrnrlat=lat_max,resolution='l')
    lon,lat=np.meshgrid(Long,Lat)
    lon,lat=m(lon,lat)
    m.drawcoastlines(linewidth=1)
    m.drawstates(linewidth=1.5)
    m.drawcountries(linewidth=2)
    m.drawparallels(np.arange(-90, 90,30), labels=[1,0,0,0], fontsize=13,linewidth=0.8)
    m.drawmeridians(np.arange(-180, 180, 45), labels=[0,0,0,1], fontsize=13,linewidth=0.8)
    cmap = plt.cm.jet_r
    norm = mpl.colors.Normalize(vmin=0, vmax=np.round(np.max(Prec[:]+1)))
    # Ref
    #cf = plt.pcolormesh(lon,lat,Var[0,0,:,:],cmap=cmap,norm=norm)
    # NMME
    cf = plt.pcolormesh(lon,lat,Var[0,0,:,:],cmap=cmap,norm=norm)
    cax=plt.axes([0.93, 0.15, 0.01,0.7])
    cbar = plt.colorbar(cf,cax=cax)
    
    return cf
```
